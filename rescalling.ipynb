```python
import os
import glob
import re
import nibabel as nib
import numpy as np

# Define a regular expression pattern to extract TIV from the file path
tiv_pattern = r'TIV_(\d+\.\d+)'

def extract_tiv_from_path(file_path):
    match = re.search(tiv_pattern, file_path)
    if match:
        tiv = float(match.group(1))
        return tiv
    else:
        return None

def rescale_gray_matter(gray_matter_img, reference_tiv):
    gray_matter_data = gray_matter_img.get_fdata()
    tiv = gray_matter_data.sum()
    scaling_factor = reference_tiv / tiv
    rescaled_gray_matter_data = gray_matter_data * scaling_factor
    return rescaled_gray_matter_data

root_folder = "/Users/siddharthdeo/Downloads/Segmented images"

reference_TIV = 1500  # Change this to your desired reference TIV

for subject_folder in glob.glob(os.path.join(root_folder, '*/*/DEFACED_NIFTI')):
    # Find all NIfTI files in the DEFACED_NIFTI folder
    nii_files = glob.glob(os.path.join(subject_folder, '*.nii.gz'))

    if len(nii_files) == 0:
        continue

    # Load the first NIfTI file (assuming there's only one per subject)
    gray_matter_img = nib.load(nii_files[0])

    # Extract TIV from the file path
    tiv = extract_tiv_from_path(nii_files[0])

    if tiv is not None:
        # Rescale gray matter volumes
        rescaled_gray_matter_data = rescale_gray_matter(gray_matter_img, reference_TIV)

        # Save the rescaled gray matter image
        output_folder = os.path.dirname(nii_files[0])
        output_file_path = os.path.join(output_folder, 'rescaled_gray_matter.nii.gz')
        rescaled_gray_matter_img = nib.Nifti1Image(rescaled_gray_matter_data, gray_matter_img.affine)
        nib.save(rescaled_gray_matter_img, output_file_path)
```
