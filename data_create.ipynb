```python
import os
import pandas as pd
from bs4 import BeautifulSoup

# Specify the directory containing XML files
directory_path = '/Users/siddharthdeo/Downloads/Segmented images/xml/'

# Initialize lists to store the extracted values and labels
all_values = []
all_labels = []

# Iterate through XML files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.xml'):
        file_path = os.path.join(directory_path, filename)

        # Parse the XML file
        with open(file_path, 'r') as xml_file:
            xml_content = xml_file.read()

        soup = BeautifulSoup(xml_content, 'xml')

        # Find all <xnat:experiment> tags with the specified attribute
        experiments = soup.find_all('xnat:experiment', attrs={'label': lambda x: x and x.startswith('WASI')})
        
        # Find all <xnat:Subject> tags and extract the 'label' attribute
        subjects = soup.find_all('xnat:Subject', label=True)

        # Iterate over experiments and match them with subjects based on position
        for i, experiment in enumerate(experiments):
            if i < len(subjects):
                label = subjects[i]['label']
                full_scale_iq_tag = experiment.find('nki:fullScaleIQ')
                
                if full_scale_iq_tag:
                    value = full_scale_iq_tag.text.strip()
                    all_values.append(value)
                    all_labels.append(label)

# Create a DataFrame from the extracted values and labels
df = pd.DataFrame({'Subject_Labels': all_labels, 'WASI_Values': all_values})

# Save the DataFrame to a CSV file
df.to_csv('wasi_values.csv', index=False)

# Print the DataFrame
print(df)
```
