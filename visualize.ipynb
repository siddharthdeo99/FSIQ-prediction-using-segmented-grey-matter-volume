```python
import nibabel as nib
import matplotlib.pyplot as plt

# Load the NIfTI image
img = nib.load('MR_1013090_500c/11/DEFACED_NIFTI/defaced_MPRAGE.nii.gz')

# Access the image data as a numpy array
data = img.get_fdata()

# Display individual slices
num_slices = data.shape[2]

# Plot a few slices
for i in range(0, num_slices, 10):  # Change the step size as needed
    plt.imshow(data[:, :, i], cmap='gray')
    plt.title(f"Slice {i}")
    plt.colorbar()
    plt.show()
```

```python
import mayavi.mlab as mlab

# Assuming data is a 3D numpy array representing the image
src = mlab.pipeline.scalar_field(data)
mlab.pipeline.volume(src, vmin=0, vmax=1)
mlab.show()
```
